```python
def unpickle(file):
    with open(file, 'rb') as fo:
        d = pickle.load(fo, encoding='bytes')
    return d
```

```python
def read_data(folder):
    x_data_temp = []
    y_data_temp = []
    x_test_data_temp = []
    y_test_data_temp = []

    for file in os.listdir(folder):
        if file.endswith(".meta") or file.endswith(".html"):
            print("Ignoring html and meta files")
        elif "test_batch" in file:
            # test data file detected. we are gonna load it separately
            test_data_temp = unpickle(folder + "//" + file)
            x_test_data_temp.append(test_data_temp[b'data'])
            y_test_data_temp.append(test_data_temp[b'labels'])
        else:
            temp_data = unpickle(folder + "//" + file)
            x_data_temp.append(temp_data[b'data'])
            y_data_temp.append(temp_data[b'labels'])
    x_data =np.array(x_data_temp)
    y_data = np.array(y_data_temp)
    x_test_data = np.array(x_test_data_temp)
    y_test_data = np.array(y_test_data_temp)
    return [x_data, y_data, x_test_data, y_test_data]
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
        featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False,  
        samplewise_std_normalization=False, zca_whitening=False,
        zca_epsilon=1e-06, rotation_range=0, width_shift_range=0.1,
        height_shift_range=0.1, shear_range=0., 
        zoom_range=0., channel_shift_range=0., fill_mode='nearest',
        cval=0., horizontal_flip=True,  
        vertical_flip=False, rescale=None,
        preprocessing_function=None,
        data_format=None,
        validation_split=0.0)
```

```python
input = Input(shape =(32,32, 1))
def convo1(input):
    x = Conv2D(8, kernel_size = (3, 3),padding ='same', activation='relu')(input)
    x = MaxPool2D(pool_size=(2,2))(x)
    x = Conv2D(16, kernel_size=(3,3),padding ='same', activation='relu')(x)
    x = MaxPool2D(pool_size=(2,2))(x)
    x = Conv2D(32, kernel_size=(3,3),padding ='same', activation='relu')(x) 
    x = MaxPool2D(pool_size=(2,2))(x)
    x = Conv2D(64, kernel_size=(3,3),padding ='same', activation='relu')(x) 
    x = MaxPool2D(pool_size=(2,2))(x)
    x = Dropout(0.2)(x)
    
    return x

layer1 = convo1(input)
layer1 = Flatten()(layer1)
layer1 = Dense(32, activation = 'relu')(layer1)
layer1 = Dense(16, activation = 'relu')(layer1)
layer1 = Dense(3, activation = 'softmax')(layer1)
model = Model(input,layer1)
```

```python
def encoder1(inp):
    net = Conv2D(32,(3,3),activation ='relu',padding='same')(inp)
    net = BatchNormalization()(net)
    net = Conv2D(32,(3,3),activation ='relu',padding='same')(net)
    net = BatchNormalization()(net)
    net= MaxPool2D(pool_size=(2,2))(net)
    net = Conv2D(64,(3,3),activation ='relu',padding='same')(net)
    net = BatchNormalization()(net)
    net = Conv2D(64,(3,3),activation ='relu',padding='same')(net)
    
    return net

def decoder1(net):
    net1 = Conv2D(32,(3,3),activation ='relu',padding='same')(encoder1(net))
    net1 = BatchNormalization()(net1)
    net1 = Conv2D(32,(3,3),activation ='relu',padding='same')(net1)
    net1 = UpSampling2D((2,2))(net1)
    decoded = Conv2D(1,(3,3),activation = 'sigmoid',padding ='same')(net1)
    
    return net1

flat =Flatten()(decoder1(convo1(input)))
out = Dense(3, activation='softmax')(flat)
```

```python
def encoder2(value):
    net11 = Conv2D(32,(3,3),activation ='relu',padding='same')(value)
    net11 = BatchNormalization()(net11)
    net11 = Conv2D(32,(3,3),activation ='relu',padding='same')(net11)
    net11 = BatchNormalization()(net11)
    net11= MaxPool2D(pool_size=(2,2))(net11)
    net11 = Conv2D(64,(3,3),activation ='relu',padding='same')(net11)
    net11 = BatchNormalization()(net11)
    net11 = Conv2D(64,(3,3),activation ='relu',padding='same')(net11)

    return net11

def decoder2(value2):
    net12 = Conv2D(32,(3,3),activation ='relu',padding='same')(encoder2(value2))
    net12 = BatchNormalization()(net12)
    net12 = Conv2D(32,(3,3),activation ='relu',padding='same')(net12)
    net12 = UpSampling2D((2,2))(net12)
    decoded1 = Conv2D(1,(3,3),activation = 'sigmoid',padding ='same')(net12)
    
    return decoded1

flat1 =Flatten()(decoder2(convo2(input)))
out1 = Dense(6, activation='softmax')(flat1)
```
